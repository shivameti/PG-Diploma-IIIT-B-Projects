```python
################  PART II   ################## 

##Merging all the state data files##
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
cwd = os.getcwd()
dirs = os.listdir( cwd )

db2=[]

for items in dirs:
    if items.find('GSVA')>0 and items.find('.csv')>0:
        x=cwd+'\\'+items
        temp_db=pd.read_csv((x), encoding = "ISO-8859-1")
        temp_db=temp_db.loc[::,['S.No.', 'Item', '2014-15']]
        temp_db['state']=items.split('-')[1]
        db2.append(temp_db)

my_gsva= pd.concat(db2)
```

```python
###Selecting the 2014-15 Per capita GDP ###
my_gsva = my_gsva.set_index('Item')  
Per_capita_GDP = my_gsva.loc[['Per Capita GSDP (Rs.)']]
Per_capita_GDP = Per_capita_GDP.sort_values('2014-15' , ascending = False, axis=0)
```

```python
## Plotting the per capita GDP for individual states ##
fig = Per_capita_GDP.plot.bar(x='state', y = '2014-15', rot=90)
plt.title("2014-15 Per capita GSDP")
plt.xticks(rotation=90)
plt.xlabel("states")
plt.ylabel("Per capita GDP")
plt.show()
```

```python
## From the Above plot##
+ Top 5 GDP are - Goa, sikkim, haryana, kerala, uttarakhand
+ Bottom 5 GDP are - Bihar, uttar pradesh, manipur, assam, jharkhand
```
